# PhytClust quick start 

This notebook shows three typical workflows:

1. **Explicit k** – `get_clusters(k=…)`
2. **Best global peaks** – `best_global(top_n=…)`
3. **One peak per resolution bin** – `best_by_resolution(num_bins=…)`

We’ll use a tiny sample tree; replace it with
your own Newick or `Bio.Phylo` object 

```python
from pathlib import Path
from Bio import Phylo
from phytclust.core import PhytClust

sample_tree = Path("/home/ganesank/project/phytclust/example/sample_tree.nwk")
tree = Phylo.read(sample_tree, "newick")

len(tree.get_terminals())
```

```python
pc = PhytClust(tree)  # defaults: no outgroup, min_size=1
pc.get_clusters(k=5)
pc.plot(outlier=True, width_scale=2, height_scale=0.3)
```

```python
pc_global = PhytClust(tree)
pc_global.best_global(top_n=2, plot_scores=True)
pc_global.plot(top_n=2, width_scale=2, height_scale=0.3)
```

```python
pc_res = PhytClust(tree)
pc_res.best_by_resolution(num_bins=3, plot_scores=True)
pc_res.plot(width_scale=2, height_scale=0.3)
```
